```python
# Set a random seed for reproducibility
np.random.seed(42)

# Calculate the number of data points for training (80%) and testing (20%)
total_samples = len(df)
train_size = int(0.8 * total_samples)

# Shuffle the dataset
shuffled_indices = np.random.permutation(total_samples)

# Split the indices into training and testing sets
train_indices = shuffled_indices[:train_size]
test_indices = shuffled_indices[train_size:]

# Create the training and testing sets
X_train, y_train = X.iloc[train_indices], y.iloc[train_indices]
X_test, y_test = X.iloc[test_indices], y.iloc[test_indices]

# Verify the shapes of the training and testing sets
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Define a function to calculate weights for Linear Regression
def weights_calculation(x_train, y_train):
    XTX = np.dot(x_train.T, x_train)
    XTY = np.dot(x_train.T, y_train)
    weights = np.linalg.solve(XTX, XTY)
    return weights

# Calculate the weights using the training data
weights = weights_calculation(X_train, y_train)

# Calculate the predicted values on both training and test data
y_train_pred = np.dot(X_train, weights)
y_test_pred = np.dot(X_test, weights)

# Calculate Mean Squared Error (MSE) for training and testing
mse_train = np.mean((y_train - y_train_pred) ** 2)
mse_test = np.mean((y_test - y_test_pred) ** 2)

# Print the MSE for training and testing
print("MSE Loss (Training):", mse_train)
print("MSE Loss (Testing):", mse_test)

# Print the final weight vector
print("Final Weight Vector (Coefficients):")
print(weights)
```
